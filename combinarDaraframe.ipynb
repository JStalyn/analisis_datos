```python
# merge
llaves_izquierda = pd.DataFrame({'key':['0','1','2','3'],
 'A':['a0','a1','a2','a3'],
 'B':['b0','b1','b2','b3']
})
llaves_derecha = pd.DataFrame({'key':['0','1','2','3'],
 'A':['a4','a5','a6','a7'],
 'B':['b4','b5','b6','b7']
})

```

```python
llaves_izquierda.merge(llaves_derecha, on ='key')
```

```python
llaves_izquierda = pd.DataFrame({'key':['0','1','2','3'],
 'A':['a0','a1','a2','a3'],
 'B':['b0','b1','b2','b3']
})
llaves_derecha = pd.DataFrame({'key2':['0','1','2','3'],
 'A':['a4','a5','a6','a7'],
 'B':['b4','b5','b6','b7']
})
```
